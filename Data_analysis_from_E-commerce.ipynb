```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Headers.2025-03-18.na-us_R (2).csv')
data
```

```python
data_filter = data[['Invoice Creation Date','Invoice Status','Payment Due Date','Invoice Status','Price Variance Amount','Quantity Variance Amount']]
data_filter
```

```python
data_filter['Quantity Variance Amount'].sum()
```

```python
data_filter['Price Variance Amount'].sum()
```

```python
# CREATED A COLUMN WITH THE TOTAL VARIANCE AMOUNT
data_filter['Shortage Amount'] = data_filter["Quantity Variance Amount"] + data_filter["Price Variance Amount"]
data_filter
```

```python
#SUM TOTAL VARIANCE AMOUNT
total_shortage = data_filter['Shortage Amount'].sum()
total_shortage
```

```python
#TRANSFOR OBJET TO DATATIME
data_filter['Invoice Creation Date'] = pd.to_datetime(data_filter['Invoice Creation Date'], dayfirst=True)
data_filter.dtypes
```

```python
#CREATED COLUMN YEAR
data_filter['Year']= data_filter['Invoice Creation Date'].dt.year
data_filter
```

```python
#SEPARATE SHORTAGE BY YEARS
annual_shortages = data_filter.groupby('Year')['Shortage Amount'].sum()
annual_shortages
```

```python
#TRANSFOR OBJET TO DATATIME
data_filter['Payment Due Date'] = pd.to_datetime(data_filter['Payment Due Date'],dayfirst=True)
data_filter.dtypes
```

```python
#CALCULATE DAY PAST DUE
data_filter['Days Past Due'] = data_filter['Payment Due Date'] - data_filter['Invoice Creation Date']
data_filter
```

```python
#TRANSFORM DAY PAST DUE TO INTERGER
data_filter['Days Past Due'] = data_filter['Days Past Due'].dt.days
data_filter.dtypes
```

```python
#CREATED A BINARI COLUMN WITH AGES SHORTAGES
data_filter['Aged Shortages'] = data_filter['Days Past Due'] > 90
data_filter
```

```python
#ADD NEW COLUMNS TO ORIGINAL DATAFRAME
data['Shortage Amount'] = data_filter['Shortage Amount']
data['Year'] = data_filter['Year']
data['Days Past Due'] = data_filter['Days Past Due']
data['Aged Shortages'] = data_filter['Aged Shortages']
data
```

```python
#GENERATION A MATRIX OF PLOTS OF ALL VARIABLES
sns.pairplot(data)
```

```python
#SELECT ONLY NUMERIC VARIABLES
data_numeric = data.select_dtypes(include=[np.number])
data_numeric
```

```python
#CREATED A MATRIX WITH ONLY NUMERIC
sns.pairplot(data_numeric)
```

```python
#CREATED A HEATMAP WITH THE CORRELATION OF NUMERIC VARIABLES
sns.heatmap(
    data_numeric.corr(method='spearman'),
    annot=True,
    cmap='coolwarm'
)
```

```python
#LINEAL REGRESSION WITH THE INVOICE AMOUNT AND SHORTAGE AMOUNT
sns.lmplot(
    x='Invoice Amount',
    y='Shortage Amount',
    data=data,
    hue='Invoice Status'
)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

axes[0].bar(data['Year'], data['Shortage Amount'])
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Total Shortage Amount')
axes[0].set_title('Annual Shortage Amounts')
axes[0].tick_params(axis='x', rotation=45)

axes[1].hist(data['Invoice Status'], bins=10)
axes[1].set_xlabel('Invoice Status')
axes[1].tick_params(axis='x', rotation=30)
axes[1].set_ylabel('Count')
axes[1].set_title('Invoice Status Distribution')


plt.tight_layout()
plt.show()
```
